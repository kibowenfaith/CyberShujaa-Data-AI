```python
#Import Libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
from google.colab import files
```

```python
# Step 1: Load the webpage
url = "https://www.scrapethissite.com/pages/forms/"
r = requests.get(url)
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
#Step 2: Find the hockey score table
table = soup.find('table', class_='table')
```

```python
#Step 3: Get Column names
column_names = [th.text.strip() for th in table.find_all('th')]
```

```python
#Step 4: Extract data rows
data_rows = []
for row in table.find_all('tr')[1:]:  # Skip header row
    data_row = [td.text.strip() for td in row.find_all('td')]
    data_rows.append(data_row)

print(data_rows[:5])
```

```python
#Step 5: Create an empty DataFrame
df = pd.DataFrame(columns=column_names)
```

```python
print(df.head())
```

```python
#Step 6: Fill the DataFrame with data
df = pd.DataFrame(data_rows, columns=column_names)
display(df.head())
```

```python
# Step 7: Export to CSV
df.to_csv("Hockey.csv", index=False)

# Download the CSV file (for Google Colab)
files.download("Hockey.csv")
```
